```python
import glob
```

```python
files = glob.glob("test/*")
```

```python
files
```

```python
file_dict = {}
for file in files:
    with open(file, 'r') as f:
        read_data = f.read()
        file_dict[int(file.split("/")[1])] = read_data
```

```python
file_dict
```

```python
X = pd.DataFrame.from_dict(file_dict, orient="Index")
```
