```python
leed_url = 'http://www.usgbc.org/projects/human-agricultural-biosciences-bldg-i?view=scorecard'
leed_req = urllib.request.Request(leed_url, headers={'User-Agent': 'Mozilla/5.0'})
leed_content = urllib.request.urlopen(leed_req).read()
leed_soup = BeautifulSoup(leed_content)
items = leed_soup.find_all( "ul", class_="sh-content")
items_list = []
for item in items:
    scores = item.find_all("li")
    for score in scores:
        items_list.append(score.get_text("|", strip=True))

items_list
```
